```python
records = []
```

```python
def print_sections(sections, level=0):
        for s in sections:
                print("-----------------------")
                record = {}
                if(len(s.text)>5):
                    sec_txt = s.text[4:]
                    sec_txt = sec_txt[:-5]
                lines = sec_txt.split("\n")
                if(len(lines) > 1):
                    for i in range(len(lines)):
                        print("\nnew : " + lines[i])
                        if("date" in record):
                            print("date déjà dans record :" + record["date"])
                            record["evenement"] = lines[i]
                            if(lines[i][:4] == "<ul>"):
                                records.append(record)
                                print("record")
                            if(lines[i][:4] == "<li>"):
                                records.append(record)
                                print("record")
                            if((lines[i][-10:] == "</ul></li>") | (lines[i][-10:] == "</li></ul>")):
                                print("on reset")
                                record = {}
                        else:
                            if((lines[i][:4] == "<li>") | (lines[i][:8] == "<ul><li>")):
                                print("on met date dans record :" + lines[i])
                                record["date"] = lines[i]
                        #si premier caractères ne sont pas <ul>, on affecte à date lines(i)
                print_sections(s.sections, level + 1)
```

```python
print_sections(p_html.sections)
```
