```python
class EchoTarget(object):
    def start(self, tag, attrib):
        print("start %s %r" % (tag, dict(attrib)))
    def end(self, tag):
        print("end %s" % tag)
    def data(self, data):
        print("data %r" % data)
    def comment(self, text):
        print("comment %s" % text)
    def close(self):
        print("close")
        return "closed!"
```

```python
parser = etree.XMLParser(target = EchoTarget())
```

```python
result = etree.XML(etree.tostring(tree.getroot(),pretty_print=True, method="html"),parser)
```
