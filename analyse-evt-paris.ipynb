```python
import pandas as pd
```

```python
df = pd.read_csv("evenements-paris.csv")
```

```python
df
```

```python
evts = []
for index, row in df.iterrows():
    evt = {}
    row["date"] = row["date"].replace("<li>", "")
    row["date"] = row["date"].replace("</li>", "")
    row["date"] = row["date"].replace("</ul>", "")
    row["date"] = row["date"].replace(".", "")
    row["date"] = row["date"].replace(" :", "")
    row["date"] = row["date"].replace(" ;", "")
    row["date"] = row["date"].replace("Vers ", "")
    row["date"] = row["date"].replace("Entre ", "")
    row["date"] = row["date"].replace("Hiver ", "")
    row["date"] = row["date"].replace("vers ", "")
    row["date"] = row["date"].replace(" et ", "-")
    row["date"] = row["date"].replace(" à ", "-")
    row["date"] = row["date"].replace(":", "")
    row["date"] = row["date"].replace(" ap J-C", "")
    row["date"] = row["date"].replace("Avant ", "")
    split = row["date"].split(" ou ")
    row["date"] = split[0]
    split = []
    split = row["date"].split(" ")
    row["date"] = split[0]
    commentaire = ""
    for i in range(len(split)):
        if(i>0):
            commentaire = commentaire + " " + split[i]
    evt["date"] = row["date"]
    evt["commentaire"] = commentaire
    split = []
    split = row["date"].split("-")
    if(split[0] == ""):
        evt["startDate"] = row["date"]
        evt["endDate"] = evt["startDate"]
    else:
        evt["startDate"] = split[0]
        evt["endDate"] = evt["startDate"]
        for i in range (len(split)):
            if(i == 1):
                evt["endDate"] = split[1]
    
    row["evenement"] = row["evenement"].replace("<li>", "")
    row["evenement"] = row["evenement"].replace("</li>", "")
    row["evenement"] = row["evenement"].replace("<ul>", "")
    row["evenement"] = row["evenement"].replace("</ul>", "")
    evt["evenement"] = row["evenement"]
    evts.append(evt)
```

```python
df_evt = pd.DataFrame(evts)
```

```python
df_evt = df_evt.drop_duplicates(subset=["evenement"])
```
